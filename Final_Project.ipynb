```python
# Code to import all libraries needed to work on the data frame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
flights_df = pd.read_csv("/content/flights_data.csv")
print(flights_df.shape)
flights_df.head()
```

# **Bar Charts:**

```python
# Code to plot data in a bar graph
sb.countplot(data = flights_df, x = 'Source');
plt.ylabel('Number of Flights', fontsize = 12)
plt.xlabel('Source', fontsize = 12)



```

```python
# Code to make colours of all the bars uniform and rotate the x axis labels at an angle of 90 degrees
base_color = sb.color_palette()[2]
sb.countplot(data = flights_df, x = 'Source', color = base_color);
plt.xticks(rotation = 30)
```

```python
# Code to arrange data in descending order
base_color = sb.color_palette()[2]
srce_order = (flights_df['Source'].value_counts().index)
sb.countplot(data = flights_df, x = 'Source', color = base_color, order = srce_order);
plt.xticks(rotation = 30)
```

```python
# Code to plot data on airlines and the number of flights from each
base_color = sb.color_palette()[3]
sb.countplot(data = flights_df, x = 'Airline', color = base_color);
```

```python
# Code to plot the data in a vertical bar graph
base_color = sb.color_palette()[3]
sb.countplot(data = flights_df, x = 'Airline', color = base_color);
plt.xticks(rotation = 90)
```

```python
#Code to plot the data in a horizontal bar graph
base_color = sb.color_palette()[3]
sb.countplot(data = flights_df, y = 'Airline', color = base_color);
plt.xticks(rotation = 90)
```

# **Count Missing Data:**

```python
#Code that finds missing data and finds the total missing data from each column
flights_df.isna().sum()
```

```python
#Code to calculate missing data and create a bar graph from the data obtained
missing_counts = flights_df.isna().sum()
base_color = sb.color_palette()[4]
sb.barplot(missing_counts.index.values, missing_counts, color = base_color)
plt.ylabel('Number of missing values', fontsize = 12)
plt.xticks(rotation = 90)
```

# **Pie Charts:**

```python
# Code to plot data in a pie chart

sorted_counts = flights_df['Destination'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False);
plt.axis('square');
plt.title('Flight Destinations')
```

```python
# Code to plot the data frame in a donut pie chart

sorted_counts = flights_df['Destination'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.5})
plt.axis('square')
plt.title('Flight Destinations')
```

# **Histograms:**

```python
#Code to plot data of duration of flighhts into a histogram

plt.hist(data = flights_df, x = 'Duration(minutes)');
```

```python
#Code to plot data of price of flights into a histogram

plt.hist(data = flights_df, x = 'Price', bins = 20);
```

```python
# Code to arrange data of prices of flights in incrementations of 1200

bins = np.arange(0, flights_df['Price'].max()+1, 1200)
plt.hist(data = flights_df, x = 'Price', bins = bins)
plt.show()
```

```python
sb.distplot(flights_df['Price']);
```

```python
sb.distplot(flights_df['Price'], kde = False);
```

```python
bin_edges = np.arange(0, flights_df['Price'].max()+1, 1200)
sb.distplot(flights_df['Price'], bins = bin_edges, kde = False, hist_kws = {'alpha' : 1});
```
